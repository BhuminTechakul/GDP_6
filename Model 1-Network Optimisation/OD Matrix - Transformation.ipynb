```python
import pandas as pd
```

```python
# transformation function - generate OD matrix for sources and potential factories

def transform_od_matrix_sf(file_path, output_path, valid_origins, valid_destinations, sheet_name='Sheet1'):
    
    od_data = pd.read_excel(file_path)
    
    filtered_data = od_data[od_data['origin_id'].isin(valid_origins) & od_data['destination_id'].isin(valid_destinations)]
    
    
    pivot_data = filtered_data.pivot_table(
        index='origin_id', 
        columns='destination_id', 
        values='total_cost', 
        aggfunc='first'
    )
    
    try:
        with pd.ExcelWriter(output_path, engine='openpyxl', mode='a') as writer:
            pivot_data.to_excel(writer, sheet_name=sheet_name)
    except FileNotFoundError:
        with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
            pivot_data.to_excel(writer, sheet_name=sheet_name)
            
# input and output - from source to factory
file_path_1 = '/Users/kouchenlu/Desktop/RiverODFull.xlsx'
output_path = '/Users/kouchenlu/Desktop/OD_SF.xlsx'
sheet_name_1 = 'OD Matrix-SF-Water' 

file_path_2 = '/Users/kouchenlu/Desktop/RoadODFull.xlsx'
output_path = '/Users/kouchenlu/Desktop/OD_SF.xlsx'
sheet_name_2 = 'OD Matrix-SF-Road' 

valid_origins = pd.read_excel('/Users/kouchenlu/Desktop/Transport Strategy.xlsx','Weighted Production')['Source'].tolist()
valid_destinations = pd.read_excel('/Users/kouchenlu/Desktop/Transport Strategy.xlsx','Town-test')['Town'].tolist() 

transform_od_matrix_sf(file_path_1, output_path, valid_origins, valid_destinations, sheet_name_1)
transform_od_matrix_sf(file_path_2, output_path, valid_origins, valid_destinations, sheet_name_2)
```
